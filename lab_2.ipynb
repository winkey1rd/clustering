```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
from scipy.spatial.distance import squareform, pdist
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet
```

```python
data = pd.read_csv('data1.txt', delimiter="   ", header=None, names=["x", "y"])
```

```python
data = data.sort_values('x')
```

```python
plt.plot(data['x'],data['y'],'r+')
```

```python
plt.scatter(data['x'],data['y'])
```

```python
kk = []
kk1 = pdist(data, 'euclidean')
kk.append(kk1)
kk2 = pdist(data, 'seuclidean', V=None)
kk.append(kk2)
kk3 = pdist(data, 'cityblock')
kk.append(kk3)
kk

#for i in k:
#    print(i)
#    print('------------------------------------------------------------')
```

```python
methods = ['single', 'complete','average']
metrics = ['euclidean', 'seuclidean', 'cityblock']
k = np.zeros( (len(methods), len(metrics)) )
Z = []
print(k)
print('------------------------------------------')
for method in methods:
    for metric in metrics:
        Z.append(linkage(data, method, metric, 'true'))
        k[methods.index(method)][metrics.index(metric)], coph_dists = cophenet(Z[methods.index(method)*3+metrics.index(metric)], kk[metrics.index(metric)])
        #print(k)
        fig = plt.figure(figsize=(25, 10))

dn = dendrogram(Z[7])
#print(k)   
#print('------------------------------------------')
```
